```python
import torch
import matplotlib.pyplot as plt
```

## Description
The next two cells load and display training data for a N-FPN trained on Quantal Response Equilibria (QRE) data. This aligns with the experiment in *What Game are We Playing?* by Kolter et al

```python
state = torch.load('NFPN_RPS_data_QRE.pth')
test_loss_hist = state['test_loss_hist']
train_loss_hist = state['train_loss_hist']
depth_host = state['depth_hist']
```

```python
fig1 = plt.figure(1)
plt.plot(train_loss_hist, linewidth=3)
plt.plot(test_loss_hist, linewidth=3)
plt.yscale('log')
plt.legend(['training loss', 'testing loss'], fontsize=15)
plt.xlabel('epochs', fontsize=15)
plt.ylabel('Mean Squared Error', fontsize=15)
# plt.savefig('N_FPN_RPS_test_loss.pdf')
plt.show()
```
